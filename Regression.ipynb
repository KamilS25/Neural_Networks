```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from sklearn import preprocessing, model_selection
```

```python
x = np.arange(0, 10.1, 0.05)
y = np.array([n**3 - 10*n**2 + 3*n + 500 for n in x])

plt.plot(x, y)
```

```python
df = pd.DataFrame([x, y])
df
```

```python
y_train = np.array(df.drop(0)).flatten()
x_train = np.array(df.drop(1)).flatten()
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(10, activation = 'relu'))
model.add(tf.keras.layers.Dense(120, activation = 'relu'))
model.add(tf.keras.layers.Dense(160, activation = 'relu'))
model.add(tf.keras.layers.Dense(1))
          
model.compile(optimizer = 'rmsprop', loss = 'mse', metrics = ['mape'])
```

```python
model.fit(tf.expand_dims(x_train, axis=-1), y_train, epochs = 700)
```

```python
pred = model.predict(x_train)
```

```python
plt.plot(x, y)
plt.plot(x_train, pred)
```
